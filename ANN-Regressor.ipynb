```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['State'] = le.fit_transform(df['State'])
df.drop(columns=['State'], inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1].values
```

```python
x
```

```python
y=df.iloc[:,4:5].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Creating ANN skeleton

regressor = Sequential() # allows to add sequential layers one after the other
regressor.add(Dense(4,activation='relu')) # input layer
regressor.add(Dense(8,activation='relu')) # hidden layer
regressor.add(Dense(10,activation='relu')) # hidden layer
regressor.add(Dense(1,activation='linear')) # output layer (linear activation function is directly propositional to input and is used for regression)
```

```python
# Compiling ANN model

regressor.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
x_train.dtype
y_train.dtype
```

```python
# training the model
regressor.fit(x_train,y_train,batch_size=10,epochs = 300)
# 10 inputs in 1 batch
```

```python
# testing the model
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
y_pred.flatten()
```

```python
# comparing the actual and predicted y values
pd.DataFrame({'Actual y_value':y_test.flatten(),'Predicted y_values':y_pred.flatten()}).head(10)
```
